```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('SINASC_RO_2019.csv')
pd.options.display.float_format = "{:,.2f}".format
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
sns.countplot(x='APGAR1', data=df);
```

```python
sns.countplot(x='APGAR5', data=df);
```

```python
sns.countplot(x='SEXO', data=df);
```

```python
# sns.countplot(x='GESTACAO', data=df);
# plt.xticks(rotation=70);
df['GESTACAO'].value_counts().sort_index().plot.bar()
```

```python
sns.countplot(x='ESTCIVMAE', data=df);
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
df['APGAR_MEDIO'] = df['APGAR5'] / df['CONSULTAS']
sns.countplot(x='APGAR_MEDIO', data=df);
plt.xticks(rotation=90);
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10,8))
fig, axes = plt.subplots(2,2)
sns.countplot(ax = axes[0,0], x='SEXO', data=df)
sns.countplot(ax = axes[0,1], x='GESTACAO', data=df)
sns.countplot(ax = axes[1,0], x='ESTCIVMAE', data=df)
sns.countplot(ax = axes[1,1], x='CONSULTAS', data=df)
# axes[0,1].set_xticklabels(rotation=45)
axes[0,1].tick_params(axis='x', rotation=70, labelsize=9)
plt.subplots_adjust(wspace=0.4, hspace=0.8)
# plt.xticks(fontsize=10)
```
